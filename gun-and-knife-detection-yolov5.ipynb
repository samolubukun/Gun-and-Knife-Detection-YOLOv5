```python
import gradio as gr
import torch
from ultralytics import YOLO
from PIL import Image
import cv2
import numpy as np
import tempfile
import os

# Load the trained model
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

def detect_objects(image_path):
    results = model(image_path)
    
    for r in results:
        # Ensure we modify class names before plotting
        for i in range(len(r.boxes.cls)):  # Loop over detected objects
            class_id = int(r.boxes.cls[i])  # Get class ID
            if r.names[class_id].lower() == "pistol":
                r.names[class_id] = "gun"  # Replace "pistol" with "gun"

        processed_image = r.plot()
        temp_image_path = tempfile.NamedTemporaryFile(suffix=".jpg", delete=False).name
        cv2.imwrite(temp_image_path, processed_image)
    
    return temp_image_path

def detect_objects_video(video_path):
    cap = cv2.VideoCapture(video_path)
    temp_video_path = tempfile.NamedTemporaryFile(suffix=".mp4", delete=False).name
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    out = cv2.VideoWriter(temp_video_path, fourcc, fps, (frame_width, frame_height))
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        results = model(frame)
        for r in results:
            # Modify detected class labels
            for i in range(len(r.boxes.cls)):  # Loop through detected objects
                class_id = int(r.boxes.cls[i])  # Get class ID
                if r.names[class_id].lower() == "pistol":
                    r.names[class_id] = "gun"  # Replace "pistol" with "gun"

            frame = r.plot()  # Apply updated labels
        
        out.write(frame)
    
    cap.release()
    out.release()
    return temp_video_path


with gr.Blocks() as iface:
    gr.Markdown("## Gun and Knife Detection using YOLOv5")
    gr.Markdown("Upload an image or a video to detect guns and knives.")
    
    with gr.Tab("Image Detection"):
        img_input = gr.Image(type="filepath")
        img_output = gr.Image(type="filepath")
        img_button = gr.Button("Detect Gun and Knife in Image")
        img_button.click(detect_objects, inputs=img_input, outputs=img_output)
    
    with gr.Tab("Video Detection"):
        video_input = gr.File(type="filepath")
        video_output = gr.Video(height=360, width=640)  # Centered, reasonable size
        video_button = gr.Button("Detect Gun and Knife in Video")
        video_button.click(detect_objects_video, inputs=video_input, outputs=video_output)

if __name__ == "__main__":
    iface.launch()
```
